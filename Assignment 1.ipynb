```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
import requests 
from StringIO import StringIO

#Describe the “Romney” data using the describe function 

data1 = requests.get('http://elections.huffingtonpost.com/pollster/2012-general-election-romney-vs-obama.csv').text 
output_data = StringIO(data1)
election_data = pd.read_csv(output_data)
election_data.info()
election_data['Romney'].describe()
```

```python
#Pollster(s) on which Romney receive the most and the lest votes

Min_Votes = election_data['Romney'].argmin()
print "Pollster where Romney has min Votes is %s" % election_data['Pollster'].iloc[Min_Votes]

Max_Votes = election_data['Romney'].argmax()
print "Pollster where Romney has max Votes is %s" % election_data['Pollster'].iloc[Max_Votes]
```

```python
#Draw a time series graph that shows Romeny’s point along pollster start date

plt.figure()
election_data.plot(x='Start Date', y='Romney').invert_xaxis()

plt.title('Time Series Graph')
plt.xlabel('Start Date')
plt.ylabel('Romney Votes')
plt.show()
```
